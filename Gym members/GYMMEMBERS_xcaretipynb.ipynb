```python
import pandas as pd
link= "https://raw.githubusercontent.com/xcaret045-bit/EstadisticaVerano2026/refs/heads/main/Gym%20members/gym_members_exercise_tracking.csv"
df=pd.read_csv(link)
df
```

```python
df.rename(columns={
    "Session_Duration (hours)": "duracion_sesion_hrs",
    "Calories_Burned": "calorias_quemadas"
}, inplace=True)
df
```

```python
#1.Establecer variables
x= df ["duracion_sesion_hrs"]
y= df ["calorias_quemadas"]
```

```python
#4. Calcula el coeficiente de correlación e interpreta el resultado
from scipy.stats import pearsonr
r, valor_p=pearsonr(x,y) #p es de prueba de hipótesis
                         # Ho: P=0 No hay correlación (el cambio en una variabole está vinculado con el cambio en otra variable)
                         # H1: P\=0 Sí hay correlación
print(f"coeficiente de correlación (r):{r:0.4f}")
print(f"valor_p:{valor_p:0.4f}")
```

Estos resultados indican que existe una correlación positiva muy fuerte entre las dos variables analizadas. Un coeficiente de correlación de $0.9081$ significa que, a medida que una variable aumenta, la otra también tiende a incrementarse de manera consistente.

```python
#5 Calcule el coeficiente de determinación e interprete el resultado. 0= r*r
print(f"coeficiente de determinación:{r**2:0.4f}")
```

sugiere que el modelo presenta un muy buen ajuste a los datos, lo que indica que existe una relación lineal fuerte entre las variables analizadas y que la variable independiente es un buen predictor de la variable dependiente. Sin embargo, aunque el modelo explica una gran proporción de la variabilidad, aún existen otros factores que pueden influir en el comportamiento de la variable de interés y que no están contemplados en este análisis.

```python
#6. Obtenga la recta de regresión ajustada y grafíquelo sobre el gráfico de dispersión.
import statsmodels.api as sm
x_constante=sm.add_constant(x)
modelo=sm.OLS(y,x_constante).fit()
y_calculada=modelo.predict(x_constante)
modelo.params
```

```python
#6. grafíquelo sobre el gráfico de dispersión.

from matplotlib.lines import lineStyles
import matplotlib.pyplot as plt
#configuración general
plt.figure(
    figsize=(6,4),  #tamaño (ancho, alto)
    dpi=100         #resolución
)
#Gráfico de dispersión
plt.scatter(
    x,y,
    marker="o",
    color="pink",
    edgecolor="purple", #color de borde
    alpha=0.9, #transparencia
    s=60, #tamaño de punto
    label="calorias_quemadas" #etiqueta
)

plt.plot(
    x,y_calculada,
    marker="o",
    color="blue",
    linewidth=2.2,
    linestyle="--",
    markersize=5,
    markerfacecolor="red",
    markeredgecolor="yellow",
    label="recta de regresión"

)

#Título
plt.title(
    "Gráfico de dispersión",
    fontsize=12, #tamaño de fuente
    fontweight="bold" #bold:negritas
)
#Etiquetas eje x
plt.xlabel(
    "duracion_sesion_hrs",
    fontsize=12,
)
#Etiqueta eje y
plt.ylabel(
    "calorias_quemadas",
    fontsize=12
)
#tamaño de los tricks
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

#Márgenes
plt.margins(x=0.05,y=0.05)
plt.gca().spines[["top","right"]].set_visible(False)

plt.grid(
    visible=True,
    linestyle="--",
    linewidth=0.7, #ancho de line
    alpha=0.5, #transparencia
    color="brown"
)
    #Leyenda
plt.legend(
        fontsize=10,
        loc="best",
        frameon=False
        #bbox_to_anchor=(0.5.-0.15)
)
        #Guardar imagen
plt.savefig(
            "Dispersión",
            bbox_inches="tight"
        )
```

```python
#7. Obtenga un intervalo de confianza del 95% para la pendiente de la recta de
#regresión ajustada ( 𝑏1 )
modelo.conf_int(alpha = 0.05)
```

```python
#8.Calcule los residuales y trace un nuevo gráfico de dispersión.
#Comente, ¿Parece que se verifican los supuestos?

residuales=modelo.resid

#Gráfico de dispersión
plt.scatter(
    x,residuales,
    marker="o",
    color="red",
    edgecolor="blue", #color de borde
    alpha=0.9, #transparencia
    s=80, #tamaño de punto
    label="calorias_quemadas" #etiqueta
)

#Título
plt.title(
    "Gráfico de dispersión",
    fontsize=13, #tamaño de fuente
    fontweight="bold" #bold:negritas
)
#Etiquetas eje x
plt.xlabel(
    "duracion_sesion_hrs",
    fontsize=12,
)
#Etiqueta eje y
plt.ylabel(
    "calorias_quemadas",
    fontsize=12
)
#tamaño de los tricks
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

#Márgenes
plt.margins(x=0.05,y=0.05)
plt.gca().spines[["top","right"]].set_visible(False)

plt.grid(
    visible=True,
    linestyle="--",
    linewidth=0.7, #ancho de line
    alpha=0.2, #transparencia
    color="black"
)
    #Leyenda
plt.legend(
        fontsize=10,
        loc="best",
        frameon=False
        #bbox_to_anchor=(0.5.-0.15)
)
        #Guardar imagen
plt.savefig(
            "Dispersión",
            bbox_inches="tight"
        )

plt.axhline(y=0, color="gray", linestyle="--") #linea horizontal

#Parece que siguen el supuesto de linealidad
```

```python
#9 Test de Shapiro
from scipy.stats import shapiro
estadistico, valor_p=shapiro(residuales)
print(f"Valor_p: {valor_p:0.4f}")
# Ho: Los datos siguen una distribución normal
# H1: Los datos NO siguen una distribución normal
```

Los resultados de la prueba de Shapiro-Wilk muestran que no se cumple el supuesto de normalidad. En consecuencia, si la normalidad es un requisito para el análisis realizado, puede ser necesario considerar transformaciones de los datos, utilizar métodos estadísticos no paramétricos o verificar si el tamaño de la muestra permite apoyarse en otros criterios, como el teorema del límite central.

```python
#Histograma y Q-Q plot
from scipy.stats import shapiro
import matplotlib.pyplot as plt
import seaborn as sns



# Visualización: Q-Q plot
sm.qqplot(residuales, line='s')
plt.title("Q-Q plot de residuos")
plt.show()

# Histograma
sns.histplot(residuales, kde=True)
plt.title("Histograma de residuos")
plt.show()
```

```python
#10. Test de Breusch-Pagan
from statsmodels.stats.api import het_breuschpagan
estadistico_1, valor_p_1, estadistico_2, valor_p_2 = het_breuschpagan(residuales, x_constante)
print(f"valor_p:{valor_p_1:0.4f}")
# Ho: No hay heterocedasticidad (varianza uniforme)
# H1: Hay heterocedasticidad (varianza no uniforme)
```

Dado que el valor p es menor que 0.05, se rechaza la hipótesis nula de la prueba, la cual establece que los residuos del modelo presentan varianza constante (homocedasticidad).

```python
#12. Realice una tabla ANOVA e interprete el resultado.
from statsmodels.formula.api import ols
# y ~ x
modelo_lineal=ols("calorias_quemadas~duracion_sesion_hrs", data=df).fit()
tabla_anova=sm.stats.anova_lm(modelo_lineal)
tabla_anova
```

El estadístico F = 4568.67 es muy elevado, lo que indica que la variabilidad explicada por el modelo es considerablemente mayor que la variabilidad atribuible al error.

Además, el valor p = 0.0000 (es decir, p < 0.001) permite rechazar la hipótesis nula, la cual establece que la duración de la sesión no tiene efecto sobre la variable dependiente.
